```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set autosave interval to 10 s, enable inline plots,
# and load the rpy2 IPython extension.
%autosave 10
%matplotlib inline
%load_ext rpy2.ipython
```

# First R Command

We can execute ``R`` commands. The results are returned to Python.

```python
%R mean(c(1, 2, 3, 4))
```

We can also assign the return values to Python variable.

```python
result = %R mean(c(1, 2, 3, 4))
result
```

# Pushing Variables to R

We can push variables into the R environment. They will appear with the same name there.

```python
data = np.array([1, 2, 3, 4])
%Rpush data
%R mean(data)
```

Variables can also be fed to R using the IPython R-magic ``-i`` parameter.

```python
data = np.array([1, 2, 3, 4])
%R -i data mean(data)
```

# Fetching Variables from R

Similarly, we can fetch variables from R. Their names will be created inside the Python environment.

**Caution** It is common in R to use dots in variable names, as in ``data.cleaned``. Since the dot has a specific meaning in Python, it cannot be part of a variable name. Thus, to avoid interoperability problems when interfacing between Python and R, you should stick to naming schemes without any dots, for example ``mixedCase`` or ``lowercase_with_underscores``.

```python
# Create a data.frame inside R.
%R data = data.frame(foo = c(1, 2, 3), bar=c(4, 5, 6))

# Fetch it into our Python environment. It will be automatically
# converted into a pandas DataFrame.
data_from_R = %Rget data
data_from_R
```

# Plotting Using R's ``ggplot2``

Let's first generate some data set to work with. We do as much processing in Python as possible, in this case we simply melt the DataFrame.

```python
np.random.seed(0)

data = pd.DataFrame({'Experiment': [1, 2, 3] * 100,
                     'A': np.random.normal(loc=1, scale=1, size=300),
                     'B': np.random.normal(loc=3, scale=1, size=300)})

data = data[['Experiment', 'A', 'B']]

data_melted = pd.melt(data, id_vars='Experiment')
data_melted
```

Now we are going to push the data into the R environment and invoke ggplot. Note that instead of invoking single R commands via `%R` (*line-magic*), we now invoke a whole IPython cell of R commands using `%%R` (*cell-magic*). Beware that cell-magic commands need their own private IPython cell to operate freely. :-)

```python
%%R -i data_melted
library(ggplot2)
ggplot(data_melted, aes(x=variable, y=value, fill=variable)) +
  geom_boxplot(outlier.shape=NA) +
  geom_jitter() +
  facet_wrap(~ Experiment) +
  ggtitle('Amazing ggplot2 Plot')
```

## Interactive Plot

All right, so we generated a simple plot using R, so far, so good; but still somewhat boring, isn't it? How about some interactive data manipulation that will be immediately reflected in the generated figures?

First, we need to import some additional tools and get a handle of our currently running interactive IPython shell.

```python
from IPython import get_ipython
from IPython.html.widgets import interact, fixed
ipython = get_ipython()
```

Next, we are going to create a very simple Python function which takes three parameters: the data we wish to plot, and a lower and an upper cutoff which we intend to manipulate interactively. These data are then passed to R's ggplot, and a figure shall be returned.

No embedded figure is returned if call ggplot using *line-magic* (`%R`), so we have to resort to *cell-magic* (`%%R`). But cell-magic, on the other hand, normally only works when it has a whole cell on its own: Using cell-magic inside a function definition is thus not possible via `%%R`.

The workaround I use is to invoke cell-magic anyway, by calling it directly via ``ipython.run_cell_magic()``.

```python
def plot_data(data, lower_cutoff, upper_cutoff):
    """
    Plot data via R's ggplot2.
    
    Parameters
    ----------
    data : DataFrame
      The data to plot. It must be a melted DataFrame
      containing the columns `Experiment`, `variable`
      and `value`.
    lower_cutoff : float
      The lower cutoff value. The cutoff itself will be
      included in the plotted data.
    upper_cutoff : float
      The lower cutoff value. The cutoff itself will be
      included in the plotted data.
      
    Returns
    -------
    A ggplot2 figure.

    Notes
    -----
    Please ensure that ggplot2 is loaded inside the R
    environment before invoking this function.
    
    """
    
    if lower_cutoff is not None:
        data = data.query('value >= @lower_cutoff')

    if upper_cutoff is not None:
        data = data.query('value <= @upper_cutoff')
    
    # We push the data into R.
    %Rpush data
    
    # We invoke cell-magic here, although the code does not have a separate
    # cell on its own.
    #
    # Please note that the 2nd parameter is empty. From the documentation
    # I figured that it could be used to pass arguments to the magic function,
    # as in run_cell_magic('R', '-i data', ...); but this did not seem to
    # work for me.
    ipython.run_cell_magic('R', '',
                           'ggplot(data, aes(x=variable, y=value, fill=variable)) + '
                           '  geom_boxplot(outlier.shape=NA) + '
                           '  geom_jitter() + '
                           '  facet_wrap(~ Experiment) + '
                           '  ggtitle("Interactive ggplot2 Plot")')
```

Finally, we can create the interaction widgets! Find  more information about ``interact`` at http://nbviewer.ipython.org/github/ipython/ipython/blob/master/examples/Interactive%20Widgets/Using%20Interact.ipynb.

```python
# Load ggplot2 inside R.
%R library(ggplot2)

# Interactiveply manipulate the data to be plotted.
# We do not want any sliders to manipulate the `data`
# parameter, so we wrap it with fixed().
interact(plot_data,
         data=fixed(data_melted),
         lower_cutoff=(0, 3, 0.1),
         upper_cutoff=(3, 6, 0.1))
```
